```python
import numpy as np
from scipy import stats

# Example arrays (replace with your own measurements)
# One value per subject — must be same length
before_rmse = np.array([
    0.000509, 0.001216, 0.000481, 0.001273, 0.000634,
    0.000644, 0.000610, 0.001162, 0.001452, 0.002204,
    0.001653, 0.000798, 0.001395, 0.001145, 0.000734,
    0.001219, 0.001203, 0.002235, 0.001959, 0.001292
])

after_rmse = np.array([
    0.000242, 0.000741, 0.000235, 0.000406, 0.000494,
    0.000275, 0.000354, 0.000493, 0.000502, 0.001110,
    0.001381, 0.000621, 0.000597, 0.000422, 0.000312,
    0.000469, 0.000686, 0.000670, 0.000763, 0.000441
])
print('RMSE paired t-test')
# Compute paired t-test (one-sided: after < before)
t_stat, p_value = stats.ttest_rel(before_rmse, after_rmse, alternative='greater')

print(f"t = {t_stat:.3f}, p = {p_value:.4e}")

# Optional: compute effect size (Cohen's dz)
diffs = before_rmse - after_rmse
dz = np.mean(diffs) / np.std(diffs, ddof=1)
print(f"Cohen's dz = {dz:.3f}")
```
